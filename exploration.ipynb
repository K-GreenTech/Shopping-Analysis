Required Analytical Questions:

1. What are the most popular colors by season?
For the Fall, brown is the most popular color, spring: baby blue, summer: lavender, and winter: black.

```python
shopping_data.groupby('Season')['Color'].agg(pd.Series.mode).apply(lambda x: x[0] if isinstance(x, list) else x)
```

2. What is the most popular clothing item by season?
Here are the most popular clothing item by season:
For Fall, it is both backpack and sweater, spring: running shoes, summer: shorts, and winter, it is leggings.

```python
shopping_data.groupby('Season')['Item Purchased'].agg(pd.Series.mode).apply(lambda x: x[0] if isinstance(x, list) else x)
```

3. What is the effect of promo codes on the dollar amount of purchases?
It seems like the promo code did not make much of a difference in the dollar amount of purchases and so there are less people using the promo code to make purchases.

```python
shopping_data.groupby('Promo Code Used')['Purchase Amount (USD)'].size()
```

4. When do users leave a review?
From this analysis, users seem to leave a review more when they do not use a promo code.

```python
shopping_data.pivot_table(values= 'Previous Purchases', index= 'Promo Code Used', columns='Review Rating', aggfunc= 'size')
```
